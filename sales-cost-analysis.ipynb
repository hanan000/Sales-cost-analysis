```python
import numpy as np
import pandas as pd
import glob

path = r'C:\Users\PERSONL\Desktop\New folder (5)'

filenames = glob.glob(path + "/*.csv")
print(filenames)

### Dataframe Initialization
concat_all_sheets_all_files = pd.DataFrame()
```

## Cleaning and concatenating all the datasets sheets and files.

```python
for file in filenames:
    print(file)
    df=pd.read_csv(file,encoding= 'unicode_escape',delimiter=';',index_col=False)
    a=df.iloc[11:,:]
    a.columns = a.iloc[0]
    a= a[1:]
    indx1=df.iloc[0:1,3:4].to_string(header=False, index=False)
    indx2=df.iloc[1:2,3:4].to_string(header=False, index=False)
    indx3=df.iloc[2:3,3:4].to_string(header=False, index=False)
    indx4=df.iloc[3:4,3:4].to_string(header=False, index=False)
    indx5=df.iloc[4:5,3:4].to_string(header=False, index=False)
    indx6=df.iloc[5:6,3:4].to_string(header=False, index=False)
    a['Tarih']=indx1
    a["Lokasyon"]=indx2
    a["Başlangıç Tarihi"]=indx3
    a["Bitiş Tarihi"]=indx4
    a["Maliyet Yöntemi"]=indx5
    a["KDV Durumu"]=indx6
    concat_all_sheets_single_file = pd.concat([a],sort=False)
    concat_all_sheets_all_files=concat_all_sheets_all_files.append(concat_all_sheets_single_file)


concat_all_sheets_all_files = concat_all_sheets_all_files.replace("þ", "ş", regex=True)
concat_all_sheets_all_files = concat_all_sheets_all_files.replace("ý", "ı", regex=True)
concat_all_sheets_all_files = concat_all_sheets_all_files.replace("Ý", "I", regex=True)
concat_all_sheets_all_files = concat_all_sheets_all_files.replace("Þ", "Ş", regex=True)	

concat_all_sheets_all_files.to_csv('concat_all_files_.csv',encoding='utf-8-sig',index=False)
```
